```python
main_folder = '/home/pgajo/working/incels/ECIR/results/ECIR_substitution_fixed_testing'

padding_right_modifier = 10
padding_left_modifier = 8
legend_height = 1.4
import os
import glob
import pandas as pd

# Create an empty list to store DataFrames
dfs = []
# Iterate over all subfolders in the main folder
for root, dirs, files in os.walk(main_folder):
    # Iterate over all CSV files in the subfolder
    for file in glob.glob(os.path.join(root, '*.csv')):
        # Check if the filename starts with 'ECIR'
        if os.path.basename(file).startswith('ECIR'):
            # Read the CSV file into a DataFrame
            temp_df = pd.read_csv(file)
            temp_df['dir'] = os.path.dirname(file)
            # Add the DataFrame to the list
            dfs.append(temp_df)

# display(dfs)
print('len(dfs)', len(dfs))

model_dict = {
    # mono
    'CNN': 'CNN',
    'bert-base-uncased': 'BERT base',
    'HateBERT': 'HateBERT',
    'bert-base-uncased-hatexplain': 'HateXplain',
    'incel-bert-base-uncased-10k_english': 'Incel BERT 10k',
    'incel-bert-base-uncased-100k_english': 'Incel BERT 100k',
    'incel-bert-base-uncased-1000k_english': 'Incel BERT 1M',
    'roberta-base': 'RoBERTa base',
    'incel-roberta-base-10k_english': 'Incel RoBERTa 10k',
    'incel-roberta-base-100k_english': 'Incel RoBERTa 100k',
    'incel-roberta-base-1000k_english': 'Incel RoBERTa 1M',
    # multi
    'bert-base-multilingual-cased': 'mBERT base',
    'incel-bert-base-multilingual-cased-10k_multi': 'Incel mBERT 10k',
    'incel-bert-base-multilingual-cased-100k_multi': 'Incel mBERT 100k',
    'incel-bert-base-multilingual-cased-1000k_multi': 'Incel mBERT 1M',
    'incel-bert-base-multilingual-cased-10k_english': 'Incel mBERT EN 10k',
    'incel-bert-base-multilingual-cased-100k_english': 'Incel mBERT EN 100k',
    'incel-bert-base-multilingual-cased-1000k_english': 'Incel mBERT EN 1M',
    'incel-bert-base-multilingual-cased-10k_italian': 'Incel mBERT IT 10k',
    'incel-bert-base-multilingual-cased-100k_italian': 'Incel mBERT IT 100k',
    'incel-bert-base-multilingual-cased-627k_italian': 'Incel mBERT IT 627k',
    # italian
    'umberto-commoncrawl-cased-v1': 'Umberto',
    'umberto-commoncrawl-cased-v1-627k_italian': 'Incel Umberto 627k',
    'bert_uncased_L-12_H-768_A-12_italian_alb3rt0': 'AlBERTo',
    'incel-bert_uncased_L-12_H-768_A-12_italian_alb3rt0-627k_italian': 'Incel AlBERTo 627k',
}

# Concatenate all DataFrames in the list
combined_df = pd.concat(dfs, ignore_index=True)
# combined_df.to_csv(os.path.join(main_folder,os.path.abspath(main_folder)+'_combined.csv')) # save merged df
combined_df['epoch'] = combined_df['epoch'].astype(int)

# Create a new column with the model name
combined_df['model'] = combined_df['model'].map(model_dict)
# print('display(combined_df)')
# display(combined_df)

# Create a list of columns to group by
group_by_list = ['dir',
'epoch',
'run_id',
'model',
'train_set(s)',
'train_source',
'dev_set(s)',
'dev_source',
'test_set(s)',
'test_source',
]

# Group by model and epoch to get the average val_f1 score for each combination of model and epoch
grouped_df = combined_df.groupby(group_by_list).agg({'val_f1': 'mean'})
# print('display(grouped_df)')
# display(grouped_df)

# Get the epoch with the highest average val_f1 score for each model
max_epochs = grouped_df.groupby(['dir', 'model', 'train_set(s)'])['val_f1'].idxmax()

for i,el in enumerate(max_epochs):
    print(i,el)


# Create a new dataframe with rows for each model at their respective max epochs

best_epoch_df = pd.concat([combined_df[(combined_df['dir'] == model_pair[0]) & (combined_df['epoch'] == model_pair[1]) & (combined_df['model'] == model_pair[3]) & (combined_df['train_set(s)'] == model_pair[4])] for model_pair in [max_epoch for max_epoch in max_epochs]])


# Print the new dataframe
if 'train_loss' in best_epoch_df.columns:
    best_epoch_df=best_epoch_df.drop(['train_loss'], axis=1)
if 'eval_loss' in best_epoch_df.columns:
    best_epoch_df=best_epoch_df.drop(['eval_loss'], axis=1)
if 'test_loss' in best_epoch_df.columns:
    best_epoch_df=best_epoch_df.drop(['test_loss'], axis=1)

# print('display(best_epoch_df)')
# display(best_epoch_df)

# Model Performance Comparison - Best Epoch for Average Test F1
import matplotlib.pyplot as plt
import numpy as np

# Choose the metrics to display on the bar chart
# You can modify the list below to include the desired metrics
metrics = [
        "val_f1",
        "val_prec",
        "val_rec",
        "test_f1",
        "test_prec",
        "test_rec",
        # 'test_incels_f1',
        # 'test_incels_prec',
        # 'test_incels_rec',
]

chart_metrics = ["val_f1",'test_f1']

# Group data by model and calculate the mean and standard deviation for the chosen metrics
# group_by_list = ['epoch', 'run_id']
mean_df = best_epoch_df.groupby(group_by_list)[metrics].mean()
std_df = best_epoch_df.groupby(group_by_list)[metrics].std()

# Sort the dataframes by val_f1 in descending order
mean_df = mean_df.sort_values("val_f1", ascending=True)
# top_test_f1 = mean_df['test_f1'].max()
# min_test_f1 = mean_df['test_f1'].min()
std_df = std_df.reindex(mean_df.index)

# Calculate the bar width and positions
num_metrics = len(chart_metrics)
num_models = len(mean_df.index)
bar_height = 1 / (num_metrics + 2)
bar_positions = np.arange(num_models) * 2 * bar_height/0.9

# Set a fixed bar height and calculate the required figure height
fig_height = num_models * (num_metrics * bar_height) # 0.1 is added for extra spacing
    
# Create a horizontal bar chart with error bars for standard deviation
fig, ax = plt.subplots(
    figsize=(8, fig_height),
)

metric_dict = {
    'val_f1': 'Val F1',
    'test_f1': 'Test F1',
    'val_prec': 'Val Prec',
    'test_prec': 'Test Prec',
    'val_rec': 'Val Rec',
    'test_rec': 'Test Rec',
    'test_incels_f1': 'Test F1',
    'test_incels_prec': 'Test Prec',
    'test_incels_rec': 'Test Rec',
    'run_id': 'ID',
    'epoch': 'Epoch',
    'test_f1_std': 'Test F1 St.Dev',
    'val_f1_std': 'Val F1 St.Dev',
    'model': 'Model',
}

# sort mean_df by test_f1
mean_df = mean_df.sort_values("test_f1", ascending=True)
display(mean_df)

run_id = mean_df.index[0][1]

for i, metric in enumerate(chart_metrics):
    ax.barh(
        bar_positions + i * bar_height,
        mean_df[metric],
        xerr=std_df[metric],
        height=bar_height,
        label=metric,
        capsize=5,
        color=['#00bfc4', '#f8766d'][i]
    )
# print(mean_df.index)

# # Remove the default yticklabels
# ax.set_yticklabels([])

# Customize the chart
ax.set_yticks(bar_positions + bar_height/2)
y_labels = [index[2] for index in mean_df.index]
ax.set_yticklabels(y_labels)  # Set y labels
ax.set_ylabel("Model", fontsize=9)
ax.set_xlabel("Score", fontsize=9)
plt.yticks(rotation=0)

# Calculate the minimum and maximum x values considering the error bars (std deviation)

print(mean_df['val_f1'].min())
print(mean_df['val_f1'].idxmin())
print(std_df['val_f1'].loc[mean_df['val_f1'].idxmin()])

min_x = mean_df['val_f1'].min() - std_df['val_f1'].loc[mean_df['val_f1'].idxmin()]
max_x = mean_df['val_f1'].max() + std_df['val_f1'].loc[mean_df['val_f1'].idxmax()]

# Calculate the padding (additional space) to add to the x limits
padding_right = (max_x - min_x) * padding_right_modifier
padding_left = (max_x - min_x) * padding_left_modifier

# Set the x limits
x_lim_max = max_x + padding_right
# x_lim_max = 1
# x_lim_min = min_x - padding_left
x_lim_min = 0
plt.xlim(x_lim_min, x_lim_max)

y_lim_max = num_models * 2 * bar_height/0.9
y_lim_min = -bar_height/0.9

plt.ylim(y_lim_min, y_lim_max)

# Add model labels inside the graph with black border and white background
for i, (index, row) in enumerate(mean_df.iterrows()):
    print(index)
    # print(str(index[4]))
    # id_string = 'ID:' + str(index[1])
    epoch_string = 'Epoch:' + str(index[0])
    ypos_epoch = bar_positions[i] + bar_height/2
    xpos_epoch = min_x - padding_right + (- min_x + max_x + 2*padding_right)/100 # Adjust the x position to fit the label inside the bar

    # test_f1_string = str(round(row['test_f1'], 3)) + '' + str(round(std_df['test_f1'].loc[index], 3)) + f' ({index[0]})'
    # ypos_f1_test = bar_positions[i] + bar_height
    # xpos_f1_test = x_lim_max - (- min_x + max_x + 2*padding_right)/100 # Adjust the x position to fit the label inside the bar
    
    val_f1_string = str(round(row['val_f1'], 3)) + '' + str(round(std_df['val_f1'].loc[index], 3)) + f' ({index[0]})'
    ypos_f1_val = bar_positions[i]
    xpos_f1_val = x_lim_max - 1*(- min_x + max_x + 2*padding_right)/100 # Adjust the x position to fit the label inside the bar

    test_f1_string = str(round(row['test_f1'], 3)) + '' + str(round(std_df['test_f1'].loc[index], 3)) + f' ({index[0]})'
    ypos_f1_test_incels = bar_positions[i] + bar_height
    xpos_f1_test_incels = x_lim_max - 1*(- min_x + max_x + 2*padding_right)/100 # Adjust the x position to fit the label inside the bar

    id_string = 'ID:' + str(index[1])
    ypos_id = bar_positions[i] + bar_height/2
    xpos_id = x_lim_min + 1*(- min_x + max_x + 2*padding_right)/100 # Adjust the x position to fit the label inside the bar

    # epoch text
    # ax.text(
    #     xpos_epoch,
    #     ypos_epoch,
    #     epoch_string,
    #     ha="left",
    #     va="center",
    #     fontsize=9,
    #     # bbox=dict(boxstyle="square,pad=0.3", edgecolor="black", facecolor=facecolor),
    # )

    # f1 test_incels text

    ax.text(
        xpos_f1_val,
        ypos_f1_val,
        val_f1_string,
        ha="right",
        va="center",
        fontsize=9,
    )
    ax.text(
        xpos_f1_test_incels,
        ypos_f1_test_incels,
        test_f1_string,
        ha="right",
        va="center",
        fontsize=9,
    )

    # id text

    ax.text(
        xpos_id,
        ypos_id,
        id_string,
        ha="left",
        va="center",
        fontsize=9,
    )

    # f1 test text
    # ax.text(
    #     xpos_f1_test,
    #     ypos_f1_test,
    #     test_f1_string,
    #     ha="right",
    #     va="center",
    #     fontsize=9,
    # )

h, l = ax.get_legend_handles_labels()
ph = [plt.plot([],marker="", ls="")[0]] # Canvas
handles = ph + h
labels = [f"ID: {run_id}"] + l  # Merging labels
plt.legend(handles, labels,ncol=3, loc='upper center', bbox_to_anchor=(0.5, legend_height), fontsize=9, frameon=False, columnspacing=0.5, handletextpad=0.5, handlelength=1.5, bbox_transform=ax.transAxes)
# chart_filename = '/home/pgajo/working/incels/data/datasets/English/Incels.is/ECIR.pdf'
# plt.savefig(chart_filename, bbox_inches='tight', pad_inches=0.05, dpi=300)
# plt.title(f"{model_name} - Best Epoch for Mean Test F1", fontsize=9)
plt.show()
```
