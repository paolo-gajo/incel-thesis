```python
import pandas as pd

file_path = '/home/pgajo/working/data/datasets/Italian/Il_forum_dei_brutti/IFC-22-IT_updated_test_subset_10k_cleaned.csv'

df = pd.read_csv(file_path)

# Remove rows where 'multi' is empty
df = df[df['multi'].notna()]

# Update the values based on 'multi' column
df.loc[df['multi'] == 0, ['hs', 'misogynous', 'racist']] = 0
df.loc[df['multi'] == 1, ['hs', 'misogynous', 'racist']] = [1, 1, 0]
df.loc[df['multi'] == 2, ['hs', 'misogynous', 'racist']] = [1, 0, 1]
df.loc[df['multi'] == 3, ['hs', 'misogynous', 'racist']] = [1, 1, 1]

# Set the other columns to 0 if 'multi' is not 1, 2, or 3
df.loc[~df['multi'].isin([1, 2, 3]), ['hs', 'misogynous', 'racist']] = 0

# Convert the columns to int
df[['hs', 'misogynous', 'racist', 'multi']] = df[['hs', 'misogynous', 'racist', 'multi']].astype(int)

display(df)
# Save the new dataframe with a suffix indicating the length of the dataset
new_file_path = file_path.replace('.csv', f'_len{len(df)}.csv')
df['data_type'] = 'test'
df.to_csv(new_file_path, index=False)
```
