```python
import pandas as pd

csv_filename = '/home/pgajo/working/data/metrics/3_hate_forecasting/metrics_monolingual/incel-bert-base-uncased-1000k_english/incel-bert-base-uncased-1000k_english_2023-04-18_19-36-40_metrics.csv'
df = pd.read_csv(csv_filename, index_col=None)

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

metrics_df = df[['epoch', 'val_mse', 'val_mae', 'val_r2', 'test_mse', 'test_mae', 'test_r2', 'mse_val_mean_baseline', 'mse_test_mean_baseline']]

# Set the style for the plots
plt.style.use('ggplot')

# Create a figure and axis to plot the metrics
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each metric
ax.plot(metrics_df['epoch'], metrics_df['val_mse'], label='Validation MSE', marker='o')
ax.plot(metrics_df['epoch'], metrics_df['test_mse'], label='Test MSE', marker='o')
ax.plot(metrics_df['epoch'], metrics_df['mse_val_mean_baseline'], label='Validation MSE mean baseline', marker='o')
ax.plot(metrics_df['epoch'], metrics_df['mse_test_mean_baseline'], label='Test MSE mean baseline', marker='o')

# Set axis labels and title using LaTeX font
ax.set_xlabel(r'Epoch')
ax.set_ylabel(r'Metrics')

# Add a legend
ax.legend()

# Display the plot
plt.show()
```
