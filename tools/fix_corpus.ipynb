```python
import pandas as pd

# Assuming you have already loaded the data into the dataframes corpus_df and supervised_df

# Remove all spaces and lowercase the text
def preprocess_text(text):
    text = ''.join(text.split())  # Remove all spaces
    return text.lower()

# Preprocess text and user columns
corpus_df['clean_text'] = corpus_df['text'].apply(preprocess_text)
corpus_df['clean_user'] = corpus_df['user'].apply(preprocess_text)
supervised_df['clean_text'] = supervised_df['text'].apply(preprocess_text)
supervised_df['clean_user'] = supervised_df['user'].apply(preprocess_text)

# Initialize 'data_type' column in corpus_df
corpus_df['data_type'] = 'unknown'

# Update 'data_type' column in corpus_df based on matches in supervised_df
for index, row in tqdm(supervised_df.iterrows(), total=len(supervised_df)):
    mask = (corpus_df['clean_text'] == row['clean_text']) & (corpus_df['clean_user'] == row['clean_user'])
    corpus_df.loc[mask, 'data_type'] = row['data_type']

# Drop temporary columns
corpus_df.drop(['clean_text', 'clean_user'], axis=1, inplace=True)
supervised_df.drop(['clean_text', 'clean_user'], axis=1, inplace=True)
```

```python
import pandas as pd
import re
from tqdm import tqdm

# Assuming you have already loaded the data into the dataframes corpus_df and supervised_df

# Remove punctuation and extra spaces
def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text)     # Remove extra spaces
    return text.lower().strip()

def jaccard_similarity(s1, s2):
    set1 = set(s1.split())
    set2 = set(s2.split())
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union

def update_data_type(corpus_row):
    best_match = None
    best_similarity = 0

    for _, supervised_row in supervised_df.iterrows():
        similarity = jaccard_similarity(corpus_row.clean_text, supervised_row['clean_text'])
        
        if similarity > best_similarity:
            best_similarity = similarity
            best_match = supervised_row

    # Set a threshold for the match, e.g., 0.8 similarity
    if best_similarity >= 0.95:
        return best_match['data_type']
    else:
        return corpus_row.data_type

# Preprocess text columns
corpus_df['clean_text'] = corpus_df['text'].apply(preprocess_text)
supervised_df['clean_text'] = supervised_df['text'].apply(preprocess_text)

# Update 'data_type' column in corpus_df
corpus_df['data_type'] = [update_data_type(row) for row in tqdm(corpus_df.itertuples(), total=len(corpus_df))]

# Drop temporary columns
corpus_df.drop('clean_text', axis=1, inplace=True)
supervised_df.drop('clean_text', axis=1, inplace=True)
```

```python
import pandas as pd
from fuzzywuzzy import fuzz
from tqdm import tqdm

# Assuming you have already loaded the data into the dataframes corpus_df and supervised_df

def update_data_type(corpus_row):
    best_match = None
    best_ratio = 0

    for _, supervised_row in supervised_df.iterrows():
        ratio = fuzz.token_sort_ratio(corpus_row.text, supervised_row['text'])
        
        if ratio > best_ratio:
            best_ratio = ratio
            best_match = supervised_row

    # Set a threshold for the match, e.g., 80% similarity
    if best_ratio >= 95:
        return best_match['data_type']
    else:
        return corpus_row.data_type

corpus_df['data_type'] = [update_data_type(row) for row in tqdm(corpus_df.itertuples(), total=len(corpus_df))]
```
