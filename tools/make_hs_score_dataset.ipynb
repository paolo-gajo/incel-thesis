```python
import pandas as pd

# df = pd.read_csv("/home/pgajo/working/data/datasets/English/Incels.is/IFC-22-EN_updated_hs_silver_labels_backup.csv")
df = pd.read_csv("/home/pgajo/working/data/datasets/Italian/Il_forum_dei_brutti/IFC-22-IT_updated_hs_silver_labels.csv")
df
```

```python
df['hs_predictions'].value_counts(normalize=1)
# Assuming your DataFrame is named 'df'
df['hs_score'] = None

# Group the dataframe by 'thread_id'
grouped = df.groupby('thread_id')

# Define a function to calculate the hs_score for each group
def calculate_hs_score(group):
    hs_sum = group['hs_predictions'][1:].sum()
    post_count = group.shape[0]
    hs_ratio = hs_sum / post_count
    group.iloc[0, df.columns.get_loc('hs_score')] = hs_ratio
    return group

# Apply the function to each group
df = grouped.apply(calculate_hs_score)

# Filter the dataframe to only keep the rows with the first post of each thread
first_posts = df[df['hs_score'].notnull()]
first_posts
```
