```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set the DPI and figsize
%matplotlib inline
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (8, 3)

# df = pd.read_csv("/home/pgajo/working/data/datasets/English/Incels.is/IFC-22-EN_updated_hs_scores_223k.csv")
df = pd.read_csv("/home/pgajo/working/data/datasets/Italian/Il_forum_dei_brutti/IFC-22-IT_updated.csv_hs_scores_30k.csv")
# df = pd.read_csv("/home/pgajo/working/data/datasets/English/Incels.is/IFC-22-EN_updated_hs_scores_223k_exclOP.csv")
# df = pd.read_csv("/home/pgajo/working/data/datasets/Italian/Il_forum_dei_brutti/IFC-22-IT_updated.csv_hs_scores_30k_exclOP.csv")
df['hs_score'] = df['hs_score'] * 100
mean = round(df['hs_score'].mean(), 2)
q25 = round(df['hs_score'].quantile(0.25), 2)
median = round(df['hs_score'].median(), 2)
q75 = round(df['hs_score'].quantile(0.75), 2)
std_dev = round(df['hs_score'].std(), 2)

print("Mean:", mean)
print("25% Quartile:", q25)
print("Median:", median)
print("75% Quartile:", q75)
print("Standard Deviation:", std_dev)

# Plot histogram with Matplotlib
plt.hist(df['hs_score'], bins=100, edgecolor='black', color='#00bfc4')
plt.xlabel('HS Score')
plt.ylabel('Frequency')
chart_filename = '/home/pgajo/working/images_tables/3_hate_forecasting/multilingual_setting/it_hs_score_hist_x100_incl0_thin_100bins.pdf'
plt.savefig(chart_filename, bbox_inches='tight', pad_inches=0.05, dpi=300)
plt.show()
```

```python
# Filter out 0 values from 'hs_score' column
filtered_df = df[df['hs_score'] != 0]

# Calculate statistics for non-zero values
mean = round(filtered_df['hs_score'].mean(), 2)
q25 = round(filtered_df['hs_score'].quantile(0.25), 2)
median = round(filtered_df['hs_score'].median(), 2)
q75 = round(filtered_df['hs_score'].quantile(0.75), 2)
std_dev = round(filtered_df['hs_score'].std(), 2)

print("Mean:", mean)
print("25% Quartile:", q25)
print("Median:", median)
print("75% Quartile:", q75)
print("Standard Deviation:", std_dev)

# Plot histogram with Matplotlib
plt.hist(filtered_df['hs_score'], bins='auto', edgecolor='black', color='#00bfc4')
plt.xlabel('HS Score')
plt.ylabel('Frequency')
# plt.title('Histogram of HS Scores (excluding 0 values)')
plt.show()
```
