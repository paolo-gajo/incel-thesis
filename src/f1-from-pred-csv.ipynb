```python
import pandas as pd
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

# Load CSV
df = pd.read_csv('/home/pgajo/working/incels/data/metrics/clic-it-2023-review/predictions/ifs-500-incel-mbert-predictions_m+r.csv')

# Drop rows with 'hs' as 'none'
df = df[df['hs'] != 'none']

# Initialize an empty DataFrame to store results
results_df = pd.DataFrame(columns=['Metric', 'hs', 'misogynous', 'racist', 'm+r'])

# Calculate metrics for 'hs' and 'predictions'
hs_precision = precision_score(df['hs'], df['predictions'], average='binary')
hs_recall = recall_score(df['hs'], df['predictions'], average='binary')
hs_f1 = f1_score(df['hs'], df['predictions'], average='binary')

# Calculate metrics for 'misogynous' and 'predictions'
miso_precision = precision_score(df['misogynous'], df['predictions'], average='binary')
miso_recall = recall_score(df['misogynous'], df['predictions'], average='binary')
miso_f1 = f1_score(df['misogynous'], df['predictions'], average='binary')

# Calculate metrics for 'racist' and 'predictions'
racist_precision = precision_score(df['racist'], df['predictions'], average='binary')
racist_recall = recall_score(df['racist'], df['predictions'], average='binary')
racist_f1 = f1_score(df['racist'], df['predictions'], average='binary')

# Calculate metrics for 'm+r' and 'predictions'
mr_precision = precision_score(df['m+r'], df['predictions'], average='binary')
mr_recall = recall_score(df['m+r'], df['predictions'], average='binary')
mr_f1 = f1_score(df['m+r'], df['predictions'], average='binary')

# Fill the results DataFrame
results_df = pd.DataFrame({
    'Metric': ['Precision', 'Recall', 'F1'],
    'hs': [hs_precision, hs_recall, hs_f1],
    'misogynous': [miso_precision, miso_recall, miso_f1],
    'racist': [racist_precision, racist_recall, racist_f1],
    'm+r': [mr_precision, mr_recall, mr_f1]
})

# Show the results DataFrame
display(results_df)
```
