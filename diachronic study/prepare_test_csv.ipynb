```python
import pandas as pd

# Replace 'your_local_file.csv' with the name of your CSV file
input_file = r'C:\Users\paolo\My Drive\UNI_Google_Drive\NLP_Google_Drive\incels_2022-2023\DIT_2023_THESIS\diachronic study\incelsis_terms_diachronic_study.csv'

# Read the CSV file
df = pd.read_csv(input_file, index_col=0)
df

# Transpose the DataFrame
df_transposed = df.transpose()
df_transposed
```

```python
df_transposed.columns
```

```python
df_final = pd.DataFrame(None,columns=['time','word','freq','group'])
for i,word in enumerate(df_transposed.columns):
    df_temp=pd.DataFrame(None,columns=['time','word','freq','group'])
    df_temp['freq'] = df_transposed[word]
    df_temp['time'] = df_transposed.index
    df_temp['word'] = word
    if i > 9:
        df_temp['group'] = 'lose'
    else:
        df_temp['group'] = 'gain'
    df_final = pd.concat([df_final,df_temp],ignore_index=True)
display(df_final)
```

```python
df_final['group'].value_counts()
```

```python
df_final.to_csv('incelsis_terms_diachronic_study_fixed.csv')
```
