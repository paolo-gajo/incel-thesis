```python
import pandas as pd

df = pd.read_csv('/home/pgajo/working/data/datasets/English/Incels.is/IFC-22-EN_updated.csv')
# df = pd.read_csv('/home/pgajo/working/data/datasets/Italian/Il_forum_dei_brutti/IFC-22-IT_updated.csv')
df = df.fillna('')
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
import pandas as pd

# Filter the first and second posts of each thread
first_posts = df[df['post_thread_num'] == 1]
second_posts = df[df['post_thread_num'] == 2]

# Merge the first and second posts using the 'thread_id' column
merged_posts = first_posts.merge(second_posts, on='thread_id', suffixes=('_first', '_second'))

# datetime_format = '%d/%m/%Y, %H:%M'
datetime_format = '%Y-%m-%dT%H:%M:%S%z'

# Convert the 'time_id' columns to datetime objects
merged_posts['time_id_first'] = pd.to_datetime(merged_posts['time_id_first'], format=datetime_format)
merged_posts['time_id_second'] = pd.to_datetime(merged_posts['time_id_second'], format=datetime_format)

# Calculate the time difference between the first and second posts
merged_posts['time_difference'] = (merged_posts['time_id_second'] - merged_posts['time_id_first']).dt.total_seconds()

# Calculate the mean time difference
mean_time_difference = merged_posts['time_difference'].mean()

print("Mean time difference between the first and second posts: ", mean_time_difference)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Group the threads by their ID in chunks of 100
thread_chunks = df['thread_id'].nunique() // 100
thread_ids = np.array_split(df['thread_id'].unique(), thread_chunks)

medians = []

for chunk in thread_ids:
    # Filter the first and second posts of each thread in the current chunk
    first_posts_chunk = first_posts[first_posts['thread_id'].isin(chunk)]
    second_posts_chunk = second_posts[second_posts['thread_id'].isin(chunk)]

    # Merge the first and second posts using the 'thread_id' column
    merged_posts_chunk = first_posts_chunk.merge(second_posts_chunk, on='thread_id', suffixes=('_first', '_second'))

    # Convert the 'time_id' columns to datetime objects, accounting for the timezone information
    merged_posts_chunk['time_id_first'] = pd.to_datetime(merged_posts_chunk['time_id_first'], format='%Y-%m-%dT%H:%M:%S%z')
    merged_posts_chunk['time_id_second'] = pd.to_datetime(merged_posts_chunk['time_id_second'], format='%Y-%m-%dT%H:%M:%S%z')

    # Calculate the time difference between the first and second posts
    merged_posts_chunk['time_difference'] = (merged_posts_chunk['time_id_second'] - merged_posts_chunk['time_id_first']).dt.total_seconds()

    # Calculate the median time difference for the current chunk
    median_time_difference = merged_posts_chunk['time_difference'].median()
    medians.append(median_time_difference)

# Plot the medians on a line graph
plt.plot(medians)
plt.xlabel('Thread Chunks (each chunk contains 100 threads)')
plt.ylabel('Median Time Difference (seconds)')
plt.title('Median Time Difference Between First and Second Posts Over Time')
plt.show()
```
