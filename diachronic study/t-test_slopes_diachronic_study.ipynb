```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.linear_model import LinearRegression

data = pd.read_csv(r'C:\Users\paolo\My Drive\UNI_Google_Drive\NLP_Google_Drive\incels_2022-2023\DIT_2023_THESIS\diachronic study\incelsis_terms_diachronic_study_fixed.csv')
data
```

```python
def calculate_slope(df):
    x = np.arange(1, 23).reshape(-1, 1)
    slopes = []
    
    for word in df['word'].unique():
        print('word:', word)
        word_data = df[df['word'] == word]
        print('word_data:', word_data)
        y = word_data['freq'].values.reshape(-1, 1)
        print('y: ', y)

        model = LinearRegression()
        model.fit(x, y)
        print('slope: ', model.coef_[0][0])
        slopes.append(model.coef_[0][0])
        
    return slopes

data_grouped = data.groupby('group')
```

```python
slopes_lose = calculate_slope(data_grouped.get_group('lose'))
print(slopes_lose)
slopes_gain = calculate_slope(data_grouped.get_group('gain'))
print(slopes_gain)
```

```python
t_statistic, p_value = stats.ttest_ind(slopes_lose, slopes_gain)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
# Write the results to a text file
with open('t_test_results_incelsis.txt', 'w') as f:
    f.write(f't-statistic: {t_statistic}\n')
    f.write(f'p-value: {p_value}\n')
```
